```python
import numpy as np
import pandas as pd
df = pd.DataFrame(np.random.randn(4, 3),index=['a', 'b', 'c', 'd'],columns=['one', 'two', 'three'])
df = df.reindex(['a', 'b', 'c'])
print("Original DataFrame:\n", df)
print("\n")
print('NaN replaced with 0:\n')
print(df.fillna(0))
```

```python
import pandas as pd
df = pd.DataFrame({
    'one': [10, 20, 80, 40, 50],
    'two': [60, 70, 80, 0, 10]
})
print("Original DataFrame:\n", df)
print("\nDataFrame with replaced values:\n")
print(df.replace({10: 5, 80: 31}))
```

```python

df = pd.DataFrame(np.random.randn(4, 3),index=['a', 'c', 'd', 'f'], columns=['one', 'two', 'three'])
df = df.reindex(['a', 'b', 'c', 'd', 'e', 'f'])
print("Original DataFrame:\n", df)
print("\nNaN with True fill:\n")
print(df['one'].isnull())
```

```python
df = pd.DataFrame(np.random.randn(4, 3),index=['a', 'c', 'd', 'f'],columns=['one', 'two', 'three'])
df = df.reindex(['a', 'b', 'c', 'd', 'e', 'f'])
print("Original DataFrame:\n", df)
print("\nNaN - Filled with Backward values:\n")
print(df.fillna(method='bfill'))
```
