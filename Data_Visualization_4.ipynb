```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px
#seaborn package
Xpoints=np.array([0,1,2,3,4,5,6])
Ypoints=np.array([10,80,100,120,150,180,250])
sns.lineplot(x=Xpoints,y=Ypoints)
plt.show()
```

```python
print(sns.scatterplot(data=Ypoints))
print(plt.show())
```

```python
print(sns.barplot(data=Ypoints))
print(plt.show())
```

```python
#3-D plot
X=np.array([1,2,6,8,12,14,20])
Y=np.array([3,8,1,10,12,16,18])
Fig=plt.figure(figsize=(12,8))
AX=plt.axes(projection='3d')
Plt_3d=AX.scatter3D(X,Y)
plt.show()
#plotly package
X=np.array([10,80,100,120,150,180,250])
fig=px.scatter(X)
fig.show()
```
