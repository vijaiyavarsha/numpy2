```python
import pandas as pd
df=pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]],index=['one','two','three'],columns=['a','b','c'])
print(df)
```

```python
print("Taking the inputfrom dataframe and storing in the excel file")
df.to_excel('excel_file_1.xlsx',sheet_name='new_sheet_name')
```

```python
#Second Dataframe input to another Excel file
d= pd.DataFrame([[110,210,310],[12,220,320],[310,320,330]],index=['four','five','six'],columns=['a','b','c'])
d.to_excel('excel_file_2.xlsx',sheet_name='new_sheet_name')
```

```python
#Merging two excel files inputinto third file
x=pd.read_excel('excel_file_1.xlsx')
y=pd.read_excel('excel_file_2.xlsx')
z=pd.concat([x,y])
z.to_excel('pandas_to_excel_merged.xlsx')
```

```python
#Sorting the column values
df=z.sort_values(["a"])
print(df)
df.to_excel('excel_file_4_sorted.xlsx')
```

```python
df=pd.read_excel('excel_file_4_sorted.xlsx')
print(df)
print(list(df))
print(format(len(df)))
```
