```python

scores = [78, 45, 90, 67, 52, 33, 81, 49, 60, 74, 88, 41, 55, 69, 72, 95, 38, 50, 64, 84]

average_score = sum(scores) / len(scores)
highest_score = max(scores)
lowest_score = min(scores)

print("Average Score:", average_score)
print("Highest Score:", highest_score)
print("Lowest Score:", lowest_score)

print("\nPass/Fail Status:")
for i, score in enumerate(scores, start=1):
    if score >= 50:
        print("Student", i, ":", score, "- Pass")
    else:
        print("Student", i, ":", score, "- Fail")

print("\nStudent Grades:")
for i, score in enumerate(scores, start=1):
    if score >= 85:
        grade = "A"
    elif score >= 70:
        grade = "B"
    elif score >= 60:
        grade = "C"
    elif score >= 50:
        grade = "D"
    else:
        grade = "F"

    print("Student", i, ":", score, "- Grade", grade)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step count data for 30 days
steps = np.array([
5000, 6200, 7000, 8000, 7500, 9000, 10000,
6500, 7200, 6800, 11000, 9500, 8700, 9200,
6000, 7000, 8500, 9700, 10200, 8800, 7600,
8200, 9100, 9900, 10500, 7300, 6400, 8900,
9400, 9800
])

# Compute statistics
total_steps = np.sum(steps)
average_steps = np.mean(steps)
max_steps = np.max(steps)
min_steps = np.min(steps)

print("Total Steps:", total_steps)
print("Average Steps per Day:", average_steps)
print("Maximum Steps:", max_steps)
print("Minimum Steps:", min_steps)

# Identify high activity days
threshold = 9000
high_activity_days = np.where(steps > threshold)

print("\nHigh Activity Days (steps > 9000):")
print("Days:", high_activity_days[0] + 1)
print("Steps:", steps[high_activity_days])

# Analyze trend (increase or decrease)
trend = np.diff(steps)
print("\nDaily Step Changes:", trend)

# Visualization
days = np.arange(1, 31)

plt.plot(days, steps, marker='o')
plt.title("Daily Step Count Over a Month")
plt.xlabel("Day")
plt.ylabel("Steps")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Grades of 30 students (rows = students, columns = subjects)
# Columns: Math, Science, English
grades = np.array([
[78, 85, 82], [90, 88, 84], [67, 72, 70], [88, 90, 91], [76, 79, 75],
[92, 94, 89], [81, 83, 80], [73, 70, 72], [85, 87, 86], [69, 74, 71],
[95, 93, 96], [77, 80, 78], [84, 82, 83], [91, 89, 90], [66, 68, 65],
[72, 75, 74], [87, 85, 88], [79, 81, 80], [93, 92, 94], [70, 73, 69],
[82, 84, 83], [88, 90, 87], [76, 78, 77], [85, 86, 84], [90, 91, 92],
[68, 70, 67], [74, 76, 73], [89, 88, 90], [83, 85, 82], [71, 72, 70]
])

subjects = ["Math", "Science", "English"]

# Subject-wise statistics
average = np.mean(grades, axis=0)
highest = np.max(grades, axis=0)
lowest = np.min(grades, axis=0)

print("Subject-wise Statistics")
for i in range(3):
    print(subjects[i])
    print("Average:", average[i])
    print("Highest:", highest[i])
    print("Lowest:", lowest[i])
    print()

# Overall statistics
overall_avg = np.mean(grades)
overall_max = np.max(grades)
overall_min = np.min(grades)

print("Overall Average Grade:", overall_avg)
print("Overall Highest Grade:", overall_max)
print("Overall Lowest Grade:", overall_min)

# Total and average score per student
total_scores = np.sum(grades, axis=1)
average_scores = np.mean(grades, axis=1)

# Identify top-performing student
top_student = np.argmax(total_scores)

print("\nTop Performing Student (based on total marks):")
print("Student Number:", top_student + 1)
print("Total Score:", total_scores[top_student])

# Visualization using histograms
plt.hist(grades[:,0], bins=10)
plt.title("Math Grade Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()

plt.hist(grades[:,1], bins=10)
plt.title("Science Grade Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()

plt.hist(grades[:,2], bins=10)
plt.title("English Grade Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()
```

```python
import numpy as np

# Enter impedance values (complex numbers)
# Example components: Resistor, Inductor, Capacitor
Z1 = complex(10, 0) # Resistor (10 Ω)
Z2 = complex(0, 5) # Inductor (j5 Ω)
Z3 = complex(0, -3) # Capacitor (-j3 Ω)

# Store impedances in a list
impedances = np.array([Z1, Z2, Z3])

print("Impedances:", impedances)

# ---- SERIES IMPEDANCE ----
Z_series = np.sum(impedances)

print("\nTotal Impedance in Series:", Z_series)

# ---- PARALLEL IMPEDANCE ----
Z_parallel = 1 / np.sum(1 / impedances)

print("Total Impedance in Parallel:", Z_parallel)

# ---- OPTIONAL USER INPUT ----
choice = input("\nEnter circuit type (series/parallel): ")

if choice == "series":
    print("Total Circuit Impedance:", Z_series)

elif choice == "parallel":
    print("Total Circuit Impedance:", Z_parallel)

else:
    print("Invalid choice")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Example EEG dataset
data = {
    "Participant": ["P1","P1","P1","P2","P2","P2","P3","P3","P3"],
    "Electrode": ["E1","E2","E3","E1","E2","E3","E1","E2","E3"],
    "Brainwave": [12.5, 15.2, 14.1, 13.3, 16.5, 15.8, 11.9, 14.7, 13.8]
}

# Load data into Pandas DataFrame
df = pd.DataFrame(data)

print("EEG Data:\n")
print(df)

# Statistical analysis
stats = df.groupby(["Participant","Electrode"])["Brainwave"].agg(["mean","var","max"])

print("\nStatistical Analysis (Mean, Variance, Peak):\n")
print(stats)

# Identify overall trends
participant_mean = df.groupby("Participant")["Brainwave"].mean()

print("\nAverage Brainwave Activity per Participant:\n")
print(participant_mean)

# Visualization of trends
participant_mean.plot(kind="bar", title="Average Brainwave Activity per Participant")
plt.xlabel("Participant")
plt.ylabel("Average Brainwave Value")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Given inventory data
data = {
'Product': ['Laptop', 'Smartphone', 'Tablet', 'Monitor', 'Keyboard'],
'Price': [1200, 800, 300, 150, 50],
'Quantity': [10, 20, 15, 25, 50]
}

# Create DataFrame
df = pd.DataFrame(data)

print("Inventory Data:\n")
print(df)

# Operation 1: Calculate total inventory value
df['Total_Value'] = df['Price'] * df['Quantity']
print("\nTotal Value of Each Product:\n")
print(df[['Product','Total_Value']])

# Operation 2: Average price
avg_price = df['Price'].mean()
print("\nAverage Price:", avg_price)

# Operation 3: Maximum price product
max_price_product = df.loc[df['Price'].idxmax()]
print("\nProduct with Maximum Price:\n", max_price_product)

# Operation 4: Minimum price product
min_price_product = df.loc[df['Price'].idxmin()]
print("\nProduct with Minimum Price:\n", min_price_product)

# Operation 5: Total quantity of products
total_quantity = df['Quantity'].sum()
print("\nTotal Quantity of All Products:", total_quantity)

# Plot: Price of products
plt.bar(df['Product'], df['Price'])
plt.title("Product Price Comparison")
plt.xlabel("Product")
plt.ylabel("Price")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the dataset
df = pd.read_csv("healthcare_data.csv")

# Display column names and data types
print("Column Names and Data Types:\n")
print(df.dtypes)

# 2. Calculate average age of patients
average_age = df["Age"].mean()
print("\nAverage Age of Patients:", average_age)

# 3. Analyze diagnosis distribution
diagnosis_counts = df["Diagnosis"].value_counts()

print("\nDiagnosis Distribution:\n")
print(diagnosis_counts)

# Visualization of diagnosis distribution
plt.bar(diagnosis_counts.index, diagnosis_counts.values)
plt.title("Diagnosis Distribution")
plt.xlabel("Diagnosis")
plt.ylabel("Number of Patients")
plt.xticks(rotation=45)
plt.show()
```
