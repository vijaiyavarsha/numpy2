```python
import pandas as pd
pd.set_option('display.max_columns',10)
print("Original data:\n")
df=pd.read_excel("DCE.xlsx")
print(df)
```

```python
f=df['Name'].str.strip()
f.to_excel("DCE.xlsx")

print("\nReplace value:")
k=df.fillna(method='pad')
print(k)
print("\nUpdated list\n")
display(k)
k.to_excel("DCE.xlsx")
```

```python
print("\nDrop one particular column and its values:\n")
k.drop(['Graphics'],axis=1,inplace=True)
print(k)
print("\nUpdated list..\n")
k.to_excel("DCE.xlsx")
```

```python
print("\nDrop NaN rows:\n")
df=pd.read_excel("DCE.xlsx")
print(df)
x=df.dropna()
print(x)
print("\nUpdated data list:\n")
x.to_excel("DCE.xlsx")
```

```python
print("\nReplace values:\n")
n=pd.read_excel("DCE.xlsx")
print("\nOriginal data:\n")
print(n)
y=n.replace({100:75})
print("\nReplaced dataset:\n")
print(y)
print("\nUpdated dataset:\n")
y.to_excel("DCE.xlsx")
```

```python
df=pd.read_excel("DCE.xlsx")
print(df)
print("\nExtract using isin() function:\n")
new=df['Hindi'].isin([100])
print(df[new])
```
